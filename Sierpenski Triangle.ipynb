```python
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline
def trans_1(p):
    a=p[0]
    b=p[1]
    x=0.5*a
    y=0.5*b
    return x,y
def trans_2(p):
    a=p[0]
    b=p[1]
    x=0.5*a+0.5
    y=0.5*b+0.5
    return x,y
def trans_3(p):
    a=p[0]
    b=p[1]
    x=0.5*a+1
    y=0.5*b
    return x,y
```

```python
a=0
b=0
transformation=[trans_1,trans_2,trans_3]
a1=[0]
b1=[0]
for x in range(10000):
    trans=choice(transformation)
    a,b=trans((a,b))
    a1.append(a)
    b1.append(b)
plt.rc('figure',figsize=(16,16))
plt.plot(a1,b1,'*+')
```
